## First: Predict the Sales and Profit in each country

```python
# -------------数据预处理---------------
import pandas as pd
import numpy as np
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.metrics import mean_squared_error

# 假设数据已加载为 df1
df1 = merged_df[['Order Date', 'Sub-Category', 'Sales', 'Profit']]

# 数据预处理
df1['Order Date'] = pd.to_datetime(df1['Order Date'])
df1.set_index('Order Date', inplace=True)
df1.sort_index(inplace=True)

# 定义特征和目标列
features = ['Sub-Category']
targets = ['Profit', 'Sales']

# 对数据进行分组聚合
df1 = df1.groupby([pd.Grouper(freq='D')] + features)[targets].sum().reset_index()
df1.set_index('Order Date', inplace=True)
df1.sort_index(inplace=True)
```

```python
# -------------模型训练---------------
for target in targets:
    # 划分训练集与测试集
    train = df1.loc['2011-01-01':'2016-12-31', [target] + features]
    test = df1.loc['2017-01-01':'2017-12-31', [target] + features]

    # 独热编码与列对齐
    train_dummy = pd.get_dummies(train, columns=features, dtype=int)
    test_dummy = pd.get_dummies(test, columns=features, dtype=int)
    test_dummy = test_dummy.reindex(columns=train_dummy.columns, fill_value=0)

    # 提取目标变量和外生变量
    train_target = train_dummy[target]
    train_exog = train_dummy.drop(target, axis=1)
    test_target = test_dummy[target]
    test_exog = test_dummy.drop(target, axis=1)

    # 检查并处理常数列
    constant_columns = train_exog.columns[train_exog.nunique() == 1]
    train_exog = train_exog.drop(columns=constant_columns)
    test_exog = test_exog.drop(columns=constant_columns)

    # 检查并处理缺失值
    train_target = train_target.fillna(0)
    train_exog = train_exog.fillna(0)

    # 构建 SARIMAX 模型
    model = SARIMAX(
        train_target,
        exog=train_exog,
        order=(1, 0, 1),
        seasonal_order=(1, 0, 1, 12)
    )
    result = model.fit()

    # 预测与评估
    predictions = result.get_forecast(steps=len(test), exog=test_exog).predicted_mean
    mse = mean_squared_error(test_target, predictions)

    # 输出结果
    print("-" * 50)
    print(f"目标变量：{target}")
    print(f"均方误差(MSE)：{mse:.4f}")
    print("-" * 50)
```

```python
#-------------预测未来利润及销售额---------------

# 定义要预测的子类
subcategories_to_predict = ['Accessories', 'Chairs']

unique_categories = df1['Sub-Category'].unique()

# 创建未来时间范围
future_dates = pd.date_range(start='2018-01-01', end='2018-01-05', freq='D')

# 遍历每个子类，生成外生变量并预测
for subcategory in subcategories_to_predict:
    # 创建未来外生变量
    future_exog = pd.DataFrame(index=future_dates)

    # 为每个类别创建列，并初始化为 0
    for category in unique_categories:
        future_exog[f'Sub-Category_{category}'] = 0
    
    # 设置当前子类的值为 1
    future_exog[f'Sub-Category_{subcategory}'] = 1
    
    # 确保外生变量列与训练数据一致
    future_exog = future_exog.reindex(columns=train_exog.columns, fill_value=0)
    
    # 使用模型进行预测
    future_forecast = result.get_forecast(steps=len(future_dates), exog=future_exog)
    future_predictions = future_forecast.predicted_mean
    future_confidence_intervals = future_forecast.conf_int()
    
    # 输出预测结果
    future_predictions_df = pd.DataFrame({
        'Date': future_dates,
        'Predicted': future_predictions,
        # 'Lower CI': future_confidence_intervals.iloc[:, 0],
        # 'Upper CI': future_confidence_intervals.iloc[:, 1]
    })
    print(f"子类 '{subcategory}' 的未来销售额预测结果：")
    print(future_predictions_df)
    print("-" * 50)
```

## Second: Customer type classification and churn warning

```python
# -------------数据预处理---------------
import pandas as pd
import numpy as np

#-----------加的子类数据筛选功能-------------

# 选择要分析的子类
selected_subcategory = 'Accessories'  
# 筛选指定子类的消费数据
df2 = merged_df[merged_df['Sub-Category'] == selected_subcategory][['Order Date', 'Order ID', 'Customer ID', 'Segment', 'City', 'State', 'Country', 'Sub-Category', 'Sales', 'Quantity', 'Discount', 'Profit', 'Shipping Cost', 'Order Priority']]


# 不指定子类
# df2 = merged_df[['Order Date','Order ID','Customer ID', 'Segment', 'City', 'State', 'Country','Sub-Category', 'Sales', 'Quantity', 'Discount','Profit', 'Shipping Cost', 'Order Priority']]

import pandas as pd
import numpy as np
from datetime import timedelta

# 假设数据已加载为 df2
# 1. 计算客户最后一次购买时间
df2['Order Date'] = pd.to_datetime(df2['Order Date'])
df2 = df2.sort_values('Order Date')

# 2. 计算 RFM 特征
rfm = df2.groupby('Customer ID').agg({
    'Order Date': lambda x: (df2['Order Date'].max() - x.max()).days,  # Recency
    'Order ID': 'count',  # Frequency
    'Sales': 'sum'  # Monetary
}).rename(columns={
    'Order Date': 'Recency',
    'Order ID': 'Frequency',
    'Sales': 'Monetary'
})

# 3. 计算其他行为特征
additional_features = df2.groupby('Customer ID').agg({
    'Discount': 'mean',  # 平均折扣使用
    'Quantity': 'mean',  # 平均购买数量
    'Shipping Cost': 'mean',  # 平均运费
    'Profit': 'mean'  # 平均利润贡献
}).rename(columns={
    'Discount': 'Avg_Discount',
    'Quantity': 'Avg_Quantity',
    'Shipping Cost': 'Avg_Shipping_Cost',
    'Profit': 'Avg_Profit'
})

# 4. 合并特征
rfm = rfm.merge(additional_features, left_index=True, right_index=True)

# 5. 标记流失客户（流失阈值：365天未购买）
rfm['Churn'] = np.where(rfm['Recency'] > 365, 1, 0)

# 6. 划分训练集与测试集（按时间排序后分割）
# 确保训练集数据截止到2016-12-31，测试集为2017年
# train_mask = df2['Order Date'] <= '2015-12-31'
# test_mask = df2['Order Date'] > '2015-12-31'
train_mask = df2['Order Date'] <= '2015-12-31'
test_mask = (df2['Order Date'] > '2015-12-31') & (df2['Order Date'] <= '2016-12-31')

# 获取训练集和测试集的客户ID
train_customers = df2[train_mask]['Customer ID'].unique()
test_customers = df2[test_mask]['Customer ID'].unique()

# 筛选训练集和测试集的RFM数据
train = rfm[rfm.index.isin(train_customers)]
test = rfm[rfm.index.isin(test_customers)]

# 7. 处理类别不平衡（仅保留特征和标签）
X_train = train.drop(['Churn'], axis=1)
y_train = train['Churn']
X_test = test.drop(['Churn'], axis=1)
y_test = test['Churn']

'''
SMOTE，全称是 Synthetic Minority Over-sampling Technique，是一种用于解决分类问题中类别不平衡的方法。当数据集中的正负样本比例相差较大时，模型可能会偏向多数类，导致对少数类的识别能力不足。这时候 SMOTE 通过生成合成样本来平衡数据集。
接下来，用户可能是在处理客户流失预警的问题时遇到了类别不平衡的情况。因为流失客户通常是少数，所以他们可能想用 SMOTE 来提升模型对流失客户的预测能力。
'''
# # 应用 SMOTE 进行过采样
import sys
sys.path.append("D:\Anaconda3\Lib\site-packages")
from imblearn.over_sampling import SMOTE

sm = SMOTE(random_state=42)
X_train_res, y_train_res = sm.fit_resample(X_train, y_train)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, roc_auc_score

# 训练模型
model_RF = RandomForestClassifier(
    n_estimators=200,
    max_depth=8,
    min_samples_split=10,
    class_weight='balanced',
    random_state=42
)
model_RF.fit(X_train_res, y_train_res)

# 评估模型
test_predictions = model_RF.predict(X_test)
test_probabilities = model_RF.predict_proba(X_test)[:, 1]  # 正确调用模型的 predict_proba

print(classification_report(y_test, test_predictions))
print(f"AUC-ROC: {roc_auc_score(y_test, test_probabilities)}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt


# 1. 预测所有客户流失概率
test_customers = X_test.index  # 获取测试集客户ID
probabilities = model_RF.predict_proba(X_test)[:, 1]  # 提取流失概率

# 2. 生成预测结果 DataFrame
prediction_df = pd.DataFrame({
    'Customer ID': test_customers,
    'Churn Probability': probabilities,
    'Predicted Churn': model_RF.predict(X_test)
})

# 3. 按流失概率排序（降序）
prediction_df = prediction_df.sort_values('Churn Probability', ascending=False)

# 4. 输出前10名高流失风险客户
print("高流失风险客户列表：")
print(prediction_df.head(5000))
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# 计算混淆矩阵
cm = confusion_matrix(y_test, test_predictions)

# 绘制混淆矩阵
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
            xticklabels=['Churn','Not Churn'], 
            yticklabels=['Churn','Not Churn'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

```python
from sklearn.metrics import roc_curve, auc

# 计算ROC曲线的FPR、TPR和阈值
fpr, tpr, thresholds = roc_curve(y_test, test_probabilities)
roc_auc = auc(fpr, tpr)

# 绘制ROC曲线
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (area = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc="lower right")
plt.show()
```
